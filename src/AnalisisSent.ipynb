```python
# Eliminar caracteres especiales y números

import re

def eliminar_caracteres_especiales(texto):
    texto_limpio = re.sub(r'[^a-zA-Z\s]', '', texto)
    return texto_limpio


txt_sin_car_num = eliminar_caracteres_especiales(txt)
```

```python
# Convertir a minúsculas

def convertir_a_minusculas(texto):
    return texto.lower()


txt_minus = convertir_a_minusculas(txt_sin_car_num)
```

```python
# Eliminar palabras vacías

from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('spanish'))

def eliminar_stopwords(texto):
    palabras = texto.split()
    palabras_filtradas = [palabra for palabra in palabras if palabra.lower() not in stop_words]
    return ' '.join(palabras_filtradas)

txt_sin_vac = eliminar_stopwords(txt_minus)
```

```python
# Lematización o derivación de palabras

from nltk.stem import SnowballStemmer

stemmer = SnowballStemmer('spanish')

def lematizar(texto):
    palabras = texto.split()
    palabras_lematizadas = [stemmer.stem(palabra) for palabra in palabras]
    return ' '.join(palabras_lematizadas)

texto_limpio = lematizar(txt_sin_vac)
```

```python
# Se importa la libreria de Vader para el analisis de Sentimiento

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

sid_obj = SentimentIntensityAnalyzer()
print(sid_obj.polarity_scores(texto_limpio))
#print(sid_obj.score_valence(filtered_text))
#print(sid_obj.sentiment_valence(filtered_text))
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

# Descarga el léxico de Vader
nltk.download('vader_lexicon')

# Crea una instancia del analizador de sentimientos
sid = SentimentIntensityAnalyzer()

# Texto de ejemplo (puedes reemplazarlo con tu propio texto)
texto = texto_limpio

# Obtiene la puntuación de polaridad del texto
scores = sid.polarity_scores(texto)

# Extrae las puntuaciones individuales
compound_score = scores['compound']
positive_score = scores['pos']
neutral_score = scores['neu']
negative_score = scores['neg']

# Etiquetas y valores para el gráfico
labels = ['Positivo', 'Neutral', 'Negativo']
values = [positive_score, neutral_score, negative_score]

# Crea el gráfico de barras
plt.bar(labels, values, color=['green', 'gray', 'red'])
plt.title('Análisis de Sentimientos')
plt.xlabel('Sentimiento')
plt.ylabel('Puntuación')
plt.show()
```
